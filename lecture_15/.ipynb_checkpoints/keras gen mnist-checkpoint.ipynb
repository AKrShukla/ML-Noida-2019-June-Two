```python
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
```

```python
chacha = image.load_img("../lecture_07/strange.jpg", target_size=(100, 200))
```

```python
chacha
```

```python
np_img = image.img_to_array(chacha)
```

```python
plt.imshow(np_img.astype(int))
```

```python
generator = image.ImageDataGenerator(horizontal_flip=True,
                        rotation_range=10,
                        shear_range=.2,
                        zoom_range=.5)
```

```python
batch = np.array([np_img, np_img])
```

```python
batch.shape
```

```python
gen = generator.flow(batch)
```

```python
ans = gen.next()
```

```python
for i in range(len(ans)):
    plt.figure()
    plt.imshow(ans[i]/255)
```
