```python
x = np.linspace(0, 10, 100)
```

```python
sqaure = x ** 2
cube = x**3
```

```python
plt.plot(x, sqaure)
plt.plot(x, cube)
```

```python
plt.figure()

plt.plot(x, sqaure, color="red")
# plt.plot(x, cube)

plt.xlim(-10, 10)
plt.ylim(0, 100)

plt.show()

plt.figure()

# plt.plot(x, sqaure)
plt.plot(x, cube, color="y")

plt.xlim(0, 10)
plt.ylim(0, 400)

plt.show()
```

```python
plt.figure()

plt.plot(x, sqaure, color="red", label="sqaure")
plt.plot(x, cube, color="y", label="cube")
plt.legend()
plt.xlim(0, 10)
plt.ylim(0, 100)

plt.xlabel("data range")
plt.ylabel("our outcome")
plt.title("Ye dekho hamare 10 number")

plt.show()
```
