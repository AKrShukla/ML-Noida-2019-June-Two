```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.normal(4, .1, 5)
```

```python
mean = [4, 6]
cov = [[2, 0], [0, 2]]

dist = np.random.multivariate_normal(mean, cov, 400)
```

```python
# dist
```

```python
plt.figure()
plt.scatter(dist[:, 0], dist[:, 1], s=6)
plt.xlim(0, 10)
plt.ylim(0, 10)

plt.show()
```

```python
mean1 = [4, 6]
cov1 = [[2, 0], [0, 2]]

mean2 = [8, 8]
cov2 = [[1, 0], [0, 2]]

dist1 = np.random.multivariate_normal(mean1, cov1, 400)
dist2 = np.random.multivariate_normal(mean2, cov2, 400)
```

```python
plt.figure()
plt.scatter(dist1[:, 0], dist1[:, 1], s=6)
plt.scatter(dist2[:, 0], dist2[:, 1], s=6)
plt.xlim(0, 15)
plt.ylim(0, 15)

plt.show()
```

```python
X = np.vstack([dist1, dist2])
```

```python
y1 = np.full((dist1.shape[0],1), 0)
y2 = np.full((dist2.shape[0],1), 1)
```

```python
y = np.vstack([y1, y2])
```

```python
y.shape
```

```python
data = np.hstack([y, X])
```

```python
data[:5]
```

```python
plt.figure()
plt.scatter(X[:, 0], X[:, 1], s=6, c=data[:, 0].flatten())
plt.xlim(0, 15)
plt.ylim(0, 15)

plt.show()
```
